# Solar Output by Hour

```python
# Import the required libraries
import requests
import json
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
```

```python
# First, figure out which zone in the Palisades Fire is in
# Using Latitude: 34.0467 and Longitude: -118.5464
response = requests.get(
        "https://api.electricitymap.org/v3/zone?lat=34.0467&lon=-118.5464",
        headers={
            "auth-token": f"API TOKEN"
        }
    )
data = response.json()
pp_zone = data['zoneKey']
print(pp_zone)

## What about for the Eaton Fire?
## Using Latitude: 34.190163 and Longitude: -118.131317
response = requests.get(
        "https://api.electricitymap.org/v3/zone?lat=34.190163&lon=-118.131317",
        headers={
            "auth-token": f"API TOKEN"
        }
    )
data = response.json()
ef_zone = data['zoneKey']
print(ef_zone)
```

```python
# Using https://stackoverflow.com/questions/69716406/looping-through-a-data-range-to-download-data-from-an-api to clean up api call


def get_solar_output(zone, start_date, end_date, temporal_granularity):
    fire_data = pd.DataFrame()
    
    delta = dt.timedelta(days=10)
    date_ranges = []
    temp_start_date = start_date
    while temp_start_date < end_date:
        temp_end_date = temp_start_date + delta 
        if temp_end_date > end_date:
            temp_end_date = end_date
        date_ranges.append([temp_start_date, temp_end_date])
        temp_start_date = temp_end_date + dt.timedelta(days=1)

    for start_date, end_date in date_ranges:
        start_date_str = start_date.strftime("%Y-%m-%d")
        end_date_str = end_date.strftime("%Y-%m-%d")

        response = requests.get(
            "https://api.electricitymaps.com/v3/electricity-source/solar/past-range?zone=" + zone + "&start=" + start_date_str + "T08%3A34%3A00.000Z&end=" + end_date_str + "T01%3A34%3A00.000Z&temporalGranularity=" + temporal_granularity,
            headers={
                "auth-token": f"API TOKEN"
            }
        )
        
        data = response.json()
        temp_df = pd.DataFrame(data['data'])
        fire_data = pd.concat([fire_data, temp_df])
    return fire_data
```

```python
start_date = dt.date(2024,12,7)
end_date = dt.date(2025,2,28)
delta = dt.timedelta(days=10)


palisades_fire_df = get_solar_output(pp_zone, start_date, end_date, 'hourly')
eaton_fire_df = get_solar_output(ef_zone, start_date, end_date, 'hourly')
```

```python
palisades_fire_df = palisades_fire_df.reset_index()
```

```python
palisades_fire_df['datetime'] = pd.to_datetime(palisades_fire_df['datetime']).dt.tz_localize(None)
```

```python
palisades_fire_df['percent'] = 0.0
for index, row in palisades_fire_df.iterrows():
    if row['datetime'].month == 12:
        palisades_fire_df.at[index, 'percent'] = row['value']/ladwp_dec2024_sc
    if row['datetime'].month == 1:
        palisades_fire_df.at[index, 'percent'] = row['value']/ladwp_jan2025_sc
    if row['datetime'].month == 2:
        palisades_fire_df.at[index, 'percent'] = row['value']/ladwp_feb2025_sc
```

```python
# Sources:
# https://www.geeksforgeeks.org/python/matplotlib-axes-axes-axvspan-in-python/
# https://stackoverflow.com/questions/57047711/how-do-i-display-only-every-nth-axis-label
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots()

ax.plot(palisades_fire_df['datetime'], palisades_fire_df['percent'])
ax.axvspan('2025-01-07 08:00:00', '2025-02-01 08:00:00', facecolor='r', alpha = 0.7)

ax.xaxis.set_major_locator(MultipleLocator(10))
plt.xticks(rotation=90)
plt.ylabel('Percentage of Solar Output in MW')
plt.xlabel('Day and Time')
plt.title('Solar Output in LADWP Before, During, and After the Palisades Fire')

plt.show()

```

```python
eaton_fire_df = eaton_fire_df.reset_index()
eaton_fire_df['datetime'] = pd.to_datetime(eaton_fire_df['datetime']).dt.tz_localize(None)

eaton_fire_df['percent'] = 0.0
for index, row in eaton_fire_df.iterrows():
    if row['datetime'].month == 12:
        eaton_fire_df.at[index, 'percent'] = row['value']/caiso_dec2024_sc
    if row['datetime'].month == 1:
        eaton_fire_df.at[index, 'percent'] = row['value']/caiso_jan2025_sc
    if row['datetime'].month == 2:
        eaton_fire_df.at[index, 'percent'] = row['value']/caiso_feb2025_sc
```

```python
# Sources:
# https://www.geeksforgeeks.org/python/matplotlib-axes-axes-axvspan-in-python/
# https://stackoverflow.com/questions/57047711/how-do-i-display-only-every-nth-axis-label
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots()

ax.plot(eaton_fire_df['datetime'], eaton_fire_df['percent'])
ax.axvspan('2025-01-07 08:00:00', '2025-02-01 08:00:00', facecolor='r', alpha = 0.7)

ax.xaxis.set_major_locator(MultipleLocator(10))
plt.xticks(rotation=90)
plt.ylabel('Percentage of Solar Output in MW')
plt.xlabel('Day and Time')
plt.title('Solar Output in CAISO Before, During, and After the Eaton Fire')

plt.show()
```

## What about when comparing to the output in January in the previous year?

```python
start_date = dt.date(2024,1,1)
end_date = dt.date(2024,1,31)
delta = dt.timedelta(days=10)


ladwp_jan24_df = get_solar_output(pp_zone, start_date, end_date, 'hourly')
caiso_jan24_df = get_solar_output(ef_zone, start_date, end_date, 'hourly')
```
